```python
df=pd.read_csv('all_data_clean.csv')
df.head()
```

```python
df.dtypes
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
df=df.loc[df['Order ID']!='Order ID']
```

```python
df.head()
```

```python
df['Order ID'] = df['Order ID'].astype(int)
```

```python
df.dtypes
```

```python
df['Order ID'] = df['Order ID'].astype(int)
df['Quantity Ordered'] = df['Quantity Ordered'].astype(int)
df['Price Each'] = df['Price Each'].astype(float)
```

```python
df.dtypes
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['Order Cost']= df['Quantity Ordered']*df['Price Each']
```

```python
df.head()
```

```python
df.drop(columns='Order Cost',inplace=True)
```

```python
df.head()
```

```python
df['Sale']=df['Quantity Ordered']*df['Price Each']
```

```python
df.head()
```

```python
df['City']=df['Purchase Address'].apply(lambda x: x.split(',')[1])
df.head()
```

```python
df['State']=df['Purchase Address'].apply(lambda x: x.split(',')[2][1:3])
df.head()
```

```python
df['Month']=pd.DatetimeIndex(df['Order Date']).month
df.head()
```

```python
df['Hour']=pd.DatetimeIndex(df['Order Date']).hour
df.head()
```

```python
df.to_csv('sales.csv')
```

```python
df.describe()
```

```python
df['City']=df['City']+'-'+df['State']
```

```python
df_montly_sale=df.groupby('Month').sum()
df_montly_sale
```

```python
import matplotlib.pyplot as plt
months=range(1,13)
plt.bar(months,df_montly_sale['Sale'])
plt.xlabel('Months')
plt.ylabel('Sale in USD')
plt.xticks(months)
plt.show()
```

```python
df_city_sale=df.groupby('City').sum()
df_city_sale
```
